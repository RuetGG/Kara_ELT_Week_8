```python
import sys
import nest_asyncio
nest_asyncio.apply()

sys.path.append("..")

from src.scraper import (
    create_telegram_client,
    fetch_channel_messages,
    save_messages_to_datalake,
    download_message_photo
    
)
```

```python
CHANNELS = {
    "chemed": "https://t.me/CheMed123",
    "lobelia": "https://t.me/lobelia4cosmetics",
    "tikvah": "https://t.me/tikvahpharma",
    "doctor online": "https://t.me/Thequorachannel"
}
```

```python
import asyncio

async def run_pipeline(channels):
    client = create_telegram_client()
    await client.start()

    for name, url in channels.items():
        print(f"Processing: {name}")

        channel_name, messages = await fetch_channel_messages(
            client,
            url
        )

        save_messages_to_datalake(
            channel_name=channel_name.replace(" ", "_").lower(),
            messages=messages
        )
        for msg_dict in messages:
            msg_id = msg_dict['id']
            msg = await client.get_messages(url, ids=msg_id)
            await download_message_photo(client,msg, channel_name.replace(" ", "_").lower() )

    await client.disconnect()

await run_pipeline(CHANNELS)
```
